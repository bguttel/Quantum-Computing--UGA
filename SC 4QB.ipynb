```python
import numpy as np
from qiskit import *
%matplotlib inline

#number of qubits
n=int(input("combien y a t-il de qubits ?"))
circ = QuantumCircuit(n,int(n/2))

#hadamard to the n-1 first qubits to have an equal repartition between 0 and 1 
#for i in range(0,int((n+1)/2)):
    #circ.h(i)

#C-NOT gates

#Z Plaquette
circ.cx(0,2)
circ.cx(1,2)

#X Plaquette
circ.h(3)
circ.cx(3,0)
circ.cx(3,1)

#Measurements

circ.measure(2,0)

circ.h(3)
circ.measure(3,1)

circ.draw()

#backend_sim = Aer.get_backend('qasm_simulator')

#job_sim = execute(circ, backend_sim, shots=30000)
#result_sim = job_sim.result()
#counts = result_sim.get_counts(circ)
#print(counts)

#from qiskit.visualization import plot_histogram
#plot_histogram(counts)


```

```python
backend_sim = Aer.get_backend('qasm_simulator')

job_sim = execute(circ, backend_sim, shots=1024)
result_sim = job_sim.result()
counts = result_sim.get_counts(circ)
print(counts)

from qiskit.visualization import plot_histogram
plot_histogram(counts)

#from qiskit import IBMQ
#IBMQ.load_account()
#IBMQ.providers()
#provider = IBMQ.get_provider(group='open')
#provider.backends()

#backend = provider.get_backend('ibmq_london')

#from qiskit.tools.monitor import job_monitor

#job_exp = execute(circ, backend=backend)
#job_monitor(job_exp)

#result_exp = job_exp.result()

#counts_exp = result_exp.get_counts(circ)
#plot_histogram([counts_exp,counts], legend=['Device', 'Simulator'])
```
